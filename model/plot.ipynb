```python
import pandas as pd;
import numpy as np;
import csv
import os

# For plotting
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
def plot_learning_curve(loss_record_dev, loss_record_train, title=''):
    ''' Plot learning curve of your DNN (train & dev loss) '''
    total_steps = len(loss_record_train)
    x_1 = range(total_steps)
    x_2 = x_1[::len(loss_record_train) // len(loss_record_dev)]
    figure(figsize=(2, 4))
    plt.plot(x_1, loss_record_train, c='tab:red', label='train')
    plt.plot(x_2, loss_record_dev, c='tab:cyan', label='dev')
    plt.ylim(0.6, 2.)
    plt.xlabel('Training steps')
    plt.ylabel('MSE loss')
    plt.title('Learning curve of {}'.format(title))
    plt.legend()
    plt.show()
```

```python
data = pd.read_csv("./lc_al0.100_et0.010.csv")
loss_dev = np.array(data['loss_dev'])
loss_train = np.array(data['loss_train'])
plot_learning_curve(loss_dev, loss_train)
```
